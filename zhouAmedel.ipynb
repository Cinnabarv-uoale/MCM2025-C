```python
zhou = {}
for row in a.itertuples():
    if int(row.Year) <=2004:
        continue
    
    z = Country2zhou[row.NOC]
    if '/' in z:
        z = z.split('/')[0]
        
    if z not in zhou:
        zhou[z]  =[]
    zhou[z].append([row.Year,row.Total])

print(zhou.keys())
```

```python
Y = {}

Y[2024] = {}
Y[2020] = {}
Y[2016] = {}
Y[2012] = {}
for key,value in zhou.items():
    if key == 'Global':
        continue
    for Ykey in Y.keys():
        if key not in Y[Ykey]:
            Y[Ykey][key] = 0
        for v in value:
            if v[0]==Ykey:
                Y[Ykey][key]+=v[1]

# for key,value in Y2024.items():
#     print('{ value :',value, ", name:  '",key,"'},")
```
