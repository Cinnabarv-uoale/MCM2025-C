```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

a = pd.read_csv("processed_data/medelAevent.csv")
print(a)
```

```python
# 7. 绘制3D折线图
fig_line = plt.figure(figsize=(50, 40))
ax_line = fig_line.add_subplot(111, projection='3d')
columns  = a.columns[1:-1]
print(columns)
names = a.loc[:,"Country"]

for row in a.itertuples():
    name = [row.Index]*len(columns)
    sports = range(len(columns))
    r = list(row)[2:-1]
    # print(r)
    ax_line.plot(name,sports,r)

# 设置轴标签和刻度
ax_line.set_xlabel('Country',fontsize = 20, labelpad=30)
ax_line.set_ylabel('Event',fontsize = 20, labelpad=30)
ax_line.set_zlabel('Athlete Count',fontsize = 20, labelpad=30)

ax_line.zaxis.set_tick_params(labelsize=16)

ax_line.set_yticks(range(len(columns)))
ax_line.set_yticklabels(columns)

ax_line.set_xticks(range(len(names)))
ax_line.set_xticklabels(names,fontsize = 16)


# 添加图例
# ax_line.legend(loc='upper left', bbox_to_anchor=(1.05, 1))

# 显示折线图
# plt.tight_layout()
plt.title('Correlation of the event with specific countries',fontsize = 30)
plt.show()
```
