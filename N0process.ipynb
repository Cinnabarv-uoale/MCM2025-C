```python
with open("processed_data/initialTrainData.json","r") as f:
    initialData = json.load(f)
with open("processed_data/sports.json","r") as f:
    sports = json.load(f)

N0 = {}

for sport in sports:
    N0[sport] = {}
    total = 0
    print(sport)
    for idx,sex in initialData[sport].items():
        total += len(sex)
        for key,athlete in sex.items():
            # print(athlete)
            if len(athlete) <= 0:
                print("len<=0")
                continue
            else:
                min_key = min(athlete, key=lambda k: k)
                # print(min_key)
                min_value = athlete[min_key]
                print(min_value)
                print(min_value[0])
                if min_value[0] not in N0[sport]:
                    N0[sport][min_value[0]] = 0
                N0[sport][min_value[0]] += 1
    for key in N0[sport]:
        N0[sport][key] /= total

print(len(N0))
json.dump(N0,open("processed_data/N0.json","w"))
```
